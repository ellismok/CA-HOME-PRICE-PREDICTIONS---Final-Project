### Cleaning EDD Employment Dataset

```python
pwd
```

```python
edd_results_df
```

```python
# Find the types of rows for area_type, aree
edd_results_df["area_type"].unique()
```

```python
# Grab only County from area_type
county_df = edd_results_df.loc[edd_results_df["area_type"]=="County"]
county_df
```

```python
# Remove the county from each name
county_df["area_name"] = county_df["area_name"].str.replace(" County", "")
county_df
```

```python
# Filtered LA county and seasonally_adjusted_y because of duplication 
filtered_county = county_df.loc[county_df["seasonally_adjusted_y_n"] == "N"]
filtered_county
```

```python
# Drop unneccessary columns: "Unnamed: 0", "area_type", "status_preliminary_final", "dates"
drop_filtered_county = filtered_county.drop(columns= ["year","month", "area_type", "seasonally_adjusted_y_n","status_preliminary_final"], axis = 1)
drop_filtered_county
```

```python
# format date column to mm-dd-yyyy
drop_filtered_county["date"] = drop_filtered_county["date"].apply(pd.to_datetime)
drop_filtered_county.head(100)
```

```python
drop_filtered_county.rename(columns = {"area_name": "county"},  
          inplace = True) 
drop_filtered_county
```

```python
# Create a new variable for last cleaned dataset
cleaned_edd_df = drop_filtered_county
cleaned_edd_df
```
